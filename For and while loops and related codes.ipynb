```python
tup1 = ((10, 20, 30), ("a", "b", "c"), (200, 300, 400))

# tuple unpacking
for i in range(3):
    for x, y, z in tup1:
        print(x, y, z)
```

```python
tup2 = ((2, 3, 4), ("a", "b", "c"), (22, 33, 44))

for i in range(3):
    for x, y, z in tup2:
        print(x, y, z)
```
